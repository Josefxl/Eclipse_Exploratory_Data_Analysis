As we perform EDA, we find that many tasks get repeated, such as plotting histograms of features. It's worth considering whether a function would help make our work more efficient. Sometimes it's not worth it. Other times, defining a function will help a lot.

The following code block defines a function that helps plot histograms with a particular style/format using this particular dataset.

```python
# define function that will plot a histogram for all attributes based on the same parameters

def histogram(column_name, median_num = True):
    median = round(waze_df[column_name].median(), 1)
    plt.figure(figsize = (5,2))
    hist = sns.histplot(x = waze_df[column_name])
    plt.axvline(median, color = 'black')
    plt.title(f' {column_name} Hist Plot')
    if median_num == True: 
        hist.text(0.20, 0.85, f' median = {median}', color = 'black', transform=hist.transAxes)
    else:
        print(f'median: {median}')
        
```

```python
# Histogram
histogram('drives')
```

The `drives` information follows a distribution similar to the `sessions` variable. It is right-skewed, approximately log-normal, with a median of 48. However, some drivers had over 400 drives in the last month.

#### **Retention by kilometers driven per driving day**


1. Create a new column in `df` called `km_per_driving_day`, which represents the mean distance driven per driving day for each user.

2. Call the `describe()` method on the new column.

```python
waze_df['km_per_driving_day'] = waze_df['driven_km_drives'] / waze_df['driving_days']
waze_df['km_per_driving_day'].describe()
```

The mean value is infinity, the standard deviation is NaN, and the max value is infinity. Why? 

This is the result of there being values of zero in the `driving_days` column. Pandas imputes a value of infinity in the corresponding rows of the new column because division by zero is undefined.

1. Convert these values from infinity to zero. we can use `np.inf` to refer to a value of infinity.

2. Call `describe()` on the `km_per_driving_day` column to verify that it worked.

```python
# replace infinity values with zero
waze_df['km_per_driving_day'] = waze_df['km_per_driving_day'].replace(np.inf, 0)

waze_df['km_per_driving_day'].describe()
```

The maximum value is 15,420 kilometers _per drive day_. This is physically impossible. Driving 100 km/hour for 12 hours is 1,200 km. It's unlikely many people averaged more than this each day they drove, so, for now, disregard rows where the distance in this column is greater than 1,200 km.

Plot a histogram of the new `km_per_driving_day` column, disregarding those users with values greater than 1,200 km. Each bar should be the same length and have two colors, one color representing the percent of the users in that bar that churned and the other representing the percent that were retained. This can be done by setting the `multiple` parameter of seaborn's [`histplot()`](https://seaborn.pydata.org/generated/seaborn.histplot.html) function to `fill`.

```python
# Histogram
plt.figure(figsize=(12,3))
sns.histplot(x= waze_df['km_per_driving_day'], bins=range(0,1201,20), 
             hue=waze_df['label'], multiple='fill')
plt.ylabel('percentage')
plt.title('Churn rate by km per driving day')
plt.show()
```

The churn rate tends to increase as the mean daily distance driven increases. It would be worth investigating further the reasons for long-distance users to discontinue using the app.

### **Task 3b. Handling outliers**

The box plots from the previous section indicated that many of these variables have outliers. These outliers do not seem to be data entry errors; they are present because of the right-skewed distributions.

Depending on the purpose of the data, it may be useful to impute outlying data with more reasonable values. One way of performing this imputation is to set a threshold based on a percentile of the distribution.

To practice this technique, write a function that calculates the 95th percentile of a given column, then imputes values > the 95th percentile with the value at the 95th percentile.  such as the 95th percentile of the distribution.



```python
# define function

def outlier_handler(column_name):
    threshold = round(waze_df[column_name].quantile(0.95),2)
    waze_df.loc[waze_df[column_name] > threshold, column_name] = threshold
    
    print(f'succesfully replaced outliers in {column_name} column with {threshold}')
```

Next, apply that function to the following columns:
* `sessions`
* `drives`
* `total_sessions`
* `driven_km_drives`
* `duration_minutes_drives`

```python
# write a for loop to apply the function on every required column
for i in ['sessions', 'drives', 'total_sessions', 'driven_km_drives', 'duration_minutes_drives']:
    outlier_handler(i)
```
